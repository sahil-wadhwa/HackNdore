```python
central_hospital_data = df[df['Hospital Name'] == 'Central Hospital']
sales_data = central_hospital_data.groupby('Equipment')['Selling Quantity'].sum().reset_index()
```

```python
plt.figure(figsize=(6,5))
plt.bar(sales_data['Equipment'], sales_data['Selling Quantity'])
plt.xlabel('Equipment')
plt.ylabel('Selling Quantity')
plt.title('Sales of Various Equipment by Central Hospital')
plt.xticks(rotation=45)
plt.show()
```

```python
df['Selling Date'] = pd.to_datetime(df['Selling Date'])
sales_data = df.groupby(['Selling Date', 'Equipment'])['Selling Quantity'].sum().reset_index()
sales_data_pivot = sales_data.pivot(index='Selling Date', columns='Equipment', values='Selling Quantity').fillna(0)
print(sales_data_pivot.head())
```

```python
def forecast_sales(equipment_sales, steps=30):
    # Fit the ARIMA model
    model = ARIMA(equipment_sales, order=(5, 1, 0))
    model_fit = model.fit()
    
    # Forecast future sales
    forecast = model_fit.forecast(steps=steps)
    return forecast
```

```python
forecast_steps = 30  # Number of days to forecast
forecasts = {}
for equipment in sales_data_pivot.columns:
    equipment_sales = sales_data_pivot[equipment]
    forecasts[equipment] = forecast_sales(equipment_sales, steps=forecast_steps)
```

```python
current_stock = {
    'MRI Machine': 50,
    'CT Scanner': 30,
    # Add other equipment as needed
}
```

```python
quantity_to_buy = {}
for equipment, forecast in forecasts.items():
    predicted_sales = forecast.sum()
    stock_needed = max(0, predicted_sales - current_stock.get(equipment, 0))
    quantity_to_buy[equipment] = stock_needed

# Display the quantity to buy
for equipment, quantity in quantity_to_buy.items():
    print(f"Quantity to buy for {equipment}: {quantity}")
```

```python
example_equipment = 'Ultrasound Machine'
plt.figure(figsize=(10, 6))
plt.plot(sales_data_pivot[example_equipment], label='Historical Sales')
plt.plot(pd.date_range(start=sales_data_pivot.index[-1], periods=forecast_steps, freq='D'), 
         forecasts[example_equipment], label='Forecasted Sales')
plt.xlabel('Date')
plt.ylabel('Selling Quantity')
plt.title(f'Sales Forecast for {example_equipment}')
plt.legend()
plt.show()
```

```python
def forecast_sales_linear(equipment_sales, steps=30):
    # Prepare the data for linear regression
    X = np.array(range(len(equipment_sales))).reshape(-1, 1)
    y = equipment_sales.values

    # Fit the linear regression model
    model = LinearRegression()
    model.fit(X, y)

    # Predict future sales
    future_X = np.array(range(len(equipment_sales), len(equipment_sales) + steps)).reshape(-1, 1)
    forecast = model.predict(future_X)
    
    return forecast

# Forecast sales for each equipment
forecast_steps = 30  # Number of days to forecast
forecasts = {}
for equipment in sales_data_pivot.columns:
    equipment_sales = sales_data_pivot[equipment]
    forecasts[equipment] = forecast_sales_linear(equipment_sales, steps=forecast_steps)

# Example current stock levels (replace with actual data)


# Calculate the quantity to buy
quantity_to_buy = {}
for equipment, forecast in forecasts.items():
    predicted_sales = forecast.sum()
    stock_needed = max(0, predicted_sales - current_stock.get(equipment, 0))
    quantity_to_buy[equipment] = stock_needed

# Display the quantity to buy
for equipment, quantity in quantity_to_buy.items():
    print(f"Quantity to buy for {equipment}: {quantity}")

# Plot the sales data and forecast for an example equipment
example_equipment = 'Anesthesia Machine'
plt.figure(figsize=(10, 6))
plt.plot(sales_data_pivot[example_equipment], label='Historical Sales')
plt.plot(pd.date_range(start=sales_data_pivot.index[-1], periods=forecast_steps, freq='D'), 
         forecasts[example_equipment], label='Forecasted Sales')
plt.xlabel('Date')
plt.ylabel('Selling Quantity')
plt.title(f'Sales Forecast for {example_equipment}')
plt.legend()
plt.show()
```
